```python
# 2-a)
# initial parameter value _SIR graph
N = 90  # total number of days
Nt = 40  # total number of people
S = zeros(N) ; I = zeros(N); R = zeros(N)
# zeros: Return a new array of given shape and type, filled with zeros
# https://numpy.org/doc/stable/reference/generated/numpy.zeros.html
CI = zeros(N)  # cummulative population
Np = zeros(N)  # new patient number

I[0] = 1; S[0] = 39; R[0] = 0; CI[0] = I[0]

beta = 1/3
gamma = 1/14

day = zeros(N)
day[0] = 0

for d in range(1, N):
    S[d] = S[d-1] - beta*I[d-1]*S[d-1]/Nt
    I[d] = I[d-1] + beta*I[d-1]*S[d-1]/Nt - gamma*I[d-1]
    R[d] = Nt -S[d] -I[d]
    Np[d] = beta*I[d-1]*S[d-1]/Nt
    CI[d] = CI[d-1] + Np[d]
    day[d] = d
plot(day, S, '-b', label="S")
plot(day, I, '-r', label="I")
plot(day, R, '-g', label="R")
plot(day, CI, '-k', label = "CIP")

xlabel('day')
ylabel('number of people')
title('Fig1. Total variation _SIR model')
legend()
show()
```

```python
# hw2-3-a)
from numpy import *
from pylab import *
from random import *
from numpy import random

# define Euclidean distance b/w p & q
def dist(p, q):
    tmp = sqrt((p[0]-q[0])**2+(p[1]-q[1])**2)
    return tmp

# parameter setting
N = 40
T = 150
x = zeros(N); y=zeros(N); vx = zeros(N); vy = zeros(N);
s = zeros(N);
R = zeros(N);
Np = zeros(T)
I = zeros(T);
Ic = zeros(T);
dt = 1
speed = 0.01
distance = 0.03
recovery = 25
# initial positions
for p in range(N):
    x[p] = rand()/2; y[p] = rand()/2
    theta = 2*pi*rand()
    vx[p] = speed *cos(theta)
    vy[p] = speed *sin(theta)
s[0] = 1; x[0] = 0.25; y[0] = 0.25

for t in range(1, T):
    for p in range(N):
        x[p] = x[p] +dt *vx[p]
        y[p] = y[p] + dt*vy[p]
        # consider the conservation of # of people
        # we need to implement the reflecting boundary conditions
        if x[p] >= 0.45:
            vx[p] = -abs(vx[p])
        elif x[p] <= 0.05:
            vx[p] = abs(vx[p])
        if y[p] >= 0.45:
            vy[p] = -abs(vy[p])
        elif y[p] <= 0.05:
            vy[p] = abs(vy[p])

# add the interaction components
    stmp = zeros(N)
    stmp = s
    newpatient = 0
    for p in range(N):
        for q in range(N):
            if s[p] == 1 and s[q] == 0:
                pc = [x[p], y[p]]
                qc = [x[q], y[q]]
                tmp = dist(pc, qc)
                if tmp <= distance:
                    comparing_p = random.random()  # here!!!
                    stochastic_p = random.normal(loc=0.8, scale=0.01, size=(1,1)) # here!!!
                    if float(comparing_p) <= float(stochastic_p):  # here!!!
                        stmp[q] = 1
                        newpatient = newpatient + 1
    Np[t] = newpatient
    Ic[t] = Ic[t-1] + Np[t]
    #important part = simultaneous update of the status
    s = stmp

# recovery
    for p in range(N):
        if s[p] ==1:
            R[p] = R[p] + 1
        if s[p] == 1 and R[p] > recovery:
            s[p] = 2

#plotting
    clf()
    for p in range(N):
        if s[p] == 0:
            plot(x[p], y[p], 'bo')
        elif s[p] == 1:
            I[t] = I[t]+1
            plot(x[p], y[p], 'ro')
        else:
            plot(x[p], y[p], 'go')
    axis((-0.01,0.51, -0.01, 0.51))
    text(0.45, 0.45, r'Day', fontsize = 15)
    text(0.45, 0.4, t, fontsize=15)
    pause(0.1)
```
