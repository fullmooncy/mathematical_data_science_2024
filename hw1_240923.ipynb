```python
# 2-a) Covid-19 data
# initial parameter value
N = 90  # total number of days
Nt = 40  # total number of people
S = zeros(N) ; I = zeros(N); R = zeros(N)
# zeros: Return a new array of given shape and type, filled with zeros
# https://numpy.org/doc/stable/reference/generated/numpy.zeros.html
CI = zeros(N)  # cummulative population
Np = zeros(N)  # new patient number

I[0] = 1; S[0] = 39; R[0] = 0; CI[0] = I[0]

beta = 1/3
gamma = 1/14

day = zeros(N)
day[0] = 0

for d in range(1, N):
    S[d] = S[d-1] - beta*I[d-1]*S[d-1]/Nt
    I[d] = I[d-1] + beta*I[d-1]*S[d-1]/Nt - gamma*I[d-1]
    R[d] = Nt -S[d] -I[d]
    Np[d] = beta*I[d-1]*S[d-1]/Nt
    CI[d] = CI[d-1] + Np[d]
    day[d] = d

plot(day, S, '-b', day, I, '-r', day, R, '-g')
title('Fig1. Total variation')
show()
```

```python
# cf: I != In != Cip
plot(day, CI, '-k', day, R, '-g')
title('Fig2. Cummulative Infected people(Black) and Recovered people(green)')
show()
```

```python
# new infection
plot(day, Np, '-r')
title('Fig3. Newly infected population')
```

```python
# make a function to show the graph by the beta value
def graph_presentation(b, g):
    N = 90  # total number of days
    Nt = 40  # total number of people
    S = zeros(N) ; I = zeros(N); R = zeros(N)
    # zeros: Return a new array of given shape and type, filled with zeros
    # https://numpy.org/doc/stable/reference/generated/numpy.zeros.html
    CI = zeros(N)  # cummulative population
    Np = zeros(N)  # new patient number
    
    I[0] = 1; S[0] = 39; R[0] = 0; CI[0] = I[0]
    
    beta = b
    gamma = g
    
    day = zeros(N)
    day[0] = 0
    
    for d in range(1, N):
        S[d] = S[d-1] - beta*I[d-1]*S[d-1]/Nt
        I[d] = I[d-1] + beta*I[d-1]*S[d-1]/Nt - gamma*I[d-1]
        R[d] = Nt -S[d] -I[d]
        Np[d] = beta*I[d-1]*S[d-1]/Nt
        CI[d] = CI[d-1] + Np[d]
        day[d] = d
    
    plot(day, S, '-b', day, I, '-r', day, R, '-g')
    title(f'Figure1. Total variation(beta:{beta})')
    show()

    plot(day, CI, '-k', day, R, '-g')
    title(f'Fig2. Cummulative Infected people(Black) and Recovered people(green)(beta:{beta})')
    show()

    plot(day, Np, '-r')
    title(f'Fig3. Newly infected population(beta:{beta})')
    show()

    plot(S, I)
    title(f'Fig4. The phase diagram of S and I(beta:{beta})')
    show()
    # x-axis is S and y-axix is I
```
