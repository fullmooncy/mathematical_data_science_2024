```python
# particle models 
# lecture material 9_25_2024
# particle model
from numpy import *
from pylab import *
from random import *
```

```python
# define Euclidean distance b/w p & q
def dist(p, q):
    tmp = sqrt((p[0]-q[0])**2+(p[1]-q[1])**2)
    return tmp
```

```python
# parameter setting
N = 40
T = 150
x = zeros(N); y=zeros(N); vx = zeros(N); vy = zeros(N);
s = zeros(N);
R = zeros(N);
Np = zeros(T)
I = zeros(T);
Ic = zeros(T);
SP = zeros(T)
RP = zeros(T)
IP = zeros(T)
day = zeros(T)

dt = 1
speed = 0.01
distance = 0.03
recovery = 25
# initial positions
for p in range(N):
    x[p] = rand()/2; y[p] = rand()/2
    theta = 2*pi*rand()
    vx[p] = speed *cos(theta)
    vy[p] = speed *sin(theta)
s[0] = 1; x[0] = 0.25; y[0] = 0.25
```

```python
for t in range(1, T):
    for p in range(N):
        x[p] = x[p] +dt *vx[p]
        y[p] = y[p] + dt*vy[p]
        # consider the conservation of # of people
        # we need to implement the reflecting boundary conditions
        if x[p] >= 0.45:
            vx[p] = -abs(vx[p])
        elif x[p] <= 0.05:
            vx[p] = abs(vx[p])
        if y[p] >= 0.45:
            vy[p] = -abs(vy[p])
        elif y[p] <= 0.05:
            vy[p] = abs(vy[p])

# add the interaction components
    stmp = zeros(N)
    stmp = s
    newpatient = 0
    for p in range(N):
        for q in range(N):
            if s[p] == 1 and s[q] == 0:
                pc = [x[p], y[p]]
                qc = [x[q], y[q]]
                tmp = dist(pc, qc)
                if tmp <=distance:
                    stmp[q] = 1
                    newpatient = newpatient + 1
    Np[t] = newpatient
    Ic[t] = Ic[t-1] + Np[t]
    #important part = simultaneous update of the status
    s = stmp

# recovery
    for p in range(N):
        if s[p] ==1:
            R[p] = R[p] + 1
        if s[p] == 1 and R[p] > recovery:
            s[p] = 2

    day[t] = t
    SP[t] = count_nonzero(s==0)
    IP[t] = count_nonzero(s==1)
    RP[t] = count_nonzero(s==2)
    # print(count_nonzero(s==0)+count_nonzero(s==1)+count_nonzero(s==2))
```

```python
# Plot the graphs of S; I; R versus t
plot(day, SP, '-b', label="S")
plot(day, IP, '-r', label="I")
plot(day, RP, '-y', label="R")
xlim([1, 150])

xlabel('day')
ylabel('number of people')
title('Fig5. the phase diagram of particle models')
legend()
show()
```
